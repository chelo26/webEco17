```python
# Libraries:
import pandas as pd
import numpy as np
from __future__ import division
import matplotlib.pyplot as plt
#plt.style.use('fivethirtyeight')
%pylab inline
import time
import joblib
from sklearn import linear_model 
from sklearn.model_selection import train_test_split
from sklearn import ensemble
```

```python
# FUNCTION TO ASSES THE PREDICTION
def get_TP(y_pred,y_test):
    TP=0
    for i in xrange(len(y_pred)):
        if y_pred[i]==1 and y_test[i]==y_pred[i]:
            TP+=1
    return TP
def get_TN(y_pred,y_test):
    TN=0
    for i in xrange(len(y_pred)):
        if y_pred[i]==0 and y_test[i]==y_pred[i]:
            TN+=1
    return TN
def get_FP(y_pred,y_test):
    FP=0
    for i in xrange(len(y_pred)):
        if y_pred[i]==1 and y_test[i]==0:
            FP+=1
    return FP
def get_FN(y_pred,y_test):
    FN=0
    for i in xrange(len(y_pred)):
        if y_pred[i]==0 and y_test[i]==1:
            FN+=1
    return FN

def get_precision(y_pred,y_test):
    return float(get_TP(y_pred,y_test))/(get_TP(y_pred,y_test) + get_FP(y_pred,y_test))

def get_recall(y_pred,y_test):
    return float(get_TP(y_pred,y_test)) / (get_TP(y_pred,y_test) + get_FN(y_pred,y_test))

def get_accuracy(y,x):
    return (get_TP(y,x) + get_TN(y,x)) / (get_TP(y,x) + get_TN(y,x) + get_FP(y,x) + get_FN(y,x))

def get_error_rate(y,x):
    return (get_FP(y,x) + get_FN(y,x)) / (get_TP(y,x) + get_TN(y,x) + get_FP(y,x) + get_FN(y,x))
```

```python
# Loading data:
start_time=time.time()
train_data=joblib.load('./data_for_CTR/data_1458.pkl')
val_data=joblib.load('./data_for_CTR/val_data_1458.pkl')
```

```python
# Separating training data in features and targets:
X_train=train_data[:,1:]
y_train=train_data[:,0]
# Separating validation in features and targets:
X_val=val_data[:,1:]
y_val=val_data[:,0]
```

```python
# class weights:
neg_w=np.sum(y_train)/len(y_train)
pos_w=1-neg_w
print "1 rebalace to : %0.5f "%pos_w
```

```python
#model = ensemble.RandomForestClassifier(n_estimators=20,class_weight={1:pos_w,0:neg_w})
model=linear_model.LogisticRegression(C=0.36,class_weight={1:pos_w,0:neg_w})#C=1000,class_weight={1:bal},penalty='l1'
# Add solver="sag"
```

```python
# Fitting the model
model.fit(X_train,y_train)
print "model fitted"
y_pred=model.predict(X_val)
print "predictions made"
print "score : " +str(model.score(X_val,y_val))
```
